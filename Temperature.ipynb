```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Dhunche_temp=pd.read_csv('temp_Dhunche.csv',na_values=-99.9, index_col=2)
Lete_temp=pd.read_csv('temp_Lete.csv',na_values=-99.9, index_col=2)
Lumle_temp=pd.read_csv('temp_Lumle.csv',na_values=-99.9, index_col=2)
Pohara_temp=pd.read_csv('temp_Pohara.csv',na_values=-99.9, index_col=2)
Rampur_temp=pd.read_csv('temp_Rampur.csv',na_values=-99.9, index_col=2)
#display(Dhunche_temp)
```

```python
Dhunche_temp.index=pd.to_datetime(Dhunche_temp.index)
type(Dhunche_temp.index)
```

#### Size

```python
display(len(Dhunche_temp))
display(len(Lete_temp))
display(len(Lumle_temp))
display(len(Pohara_temp))
display(len(Rampur_temp))
```

#### Mean of max and min temp

```python
Dhunche_temp['Mean temp'] = (Dhunche_temp['Tmax']+Dhunche_temp['Tmin'])/2
Lete_temp['Mean temp'] = (Lete_temp['Tmax']+Lete_temp['Tmin'])/2
Lumle_temp['Mean temp'] = (Lumle_temp['Tmax']+Lumle_temp['Tmin'])/2
Pohara_temp['Mean temp'] = (Pohara_temp['Tmax']+Pohara_temp['Tmin'])/2
Rampur_temp['Mean temp'] = (Rampur_temp['Tmax']+Rampur_temp['Tmin'])/2
#display(Lete_temp)
```

### Make new df with mean temp

```python
Dhunche_mean_temp=Dhunche_temp['Mean temp']
Dhunche_mean_temp=Dhunche_mean_temp.dropna()

Lete_mean_temp=Lete_temp['Mean temp']
Lete_mean_temp=Lete_mean_temp.dropna()

Lumle_mean_temp=Lumle_temp['Mean temp']
Lumle_mean_temp=Lumle_mean_temp.dropna()

Pohara_mean_temp=Pohara_temp['Mean temp']
Pohara_mean_temp=Pohara_mean_temp.dropna()

Rampur_mean_temp=Rampur_temp['Mean temp']
Rampur_mean_temp=Rampur_mean_temp.dropna()
#type(Dhunche_mean_temp.index)
#display(Dhunche_mean_temp)
```
